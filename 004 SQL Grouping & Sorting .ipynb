# GROUPING in SQL (GROUP BY)

```python
# GROUP BY is used to combine rows having the same values in specified columns into summary rows.

# It is mainly used with aggregate functions.
```

```python
# Aggregate Functions Used with GROUP BY

# Function    	Purpose

# COUNT()	        Counts rows
# SUM()	        Adds values
# AVG()	        Finds average
# MAX()	        Finds maximum value
# MIN()	        Finds minimum value
```

```python
# SELECT column_name, aggregate_function(column_name)
# FROM table_name
# GROUP BY column_name;
```

## Important Rules of GROUP BY

```python
#1. Every non-aggregated column in SELECT must be inside GROUP BY
#2. WHERE filters rows BEFORE grouping
#3. HAVING filters groups AFTER grouping
```

## Multiple Column GROUP BY

```python
# SELECT city, customer, SUM(amount)
# FROM orders
# GROUP BY city, customer;

# Groups data using both columns.
```

# SORTING in SQL (ORDER BY)

```python
# ORDER BY is used to arrange result rows in a specific order
```

```python
# Syntax

# SELECT *
# FROM table_name
# ORDER BY column_name ASC/DESC;
```

## Sorting by Multiple Columns

```python
# SELECT *
# FROM orders
# ORDER BY city ASC, amount DESC;

# Meaning:

# Sort by city
# Inside same city, sort by amount descending
```

```python
# ORDER BY using Column Number


# SELECT customer, amount
# FROM orders
# ORDER BY 2 DESC;

# 2 means second selected column (amount)
```

# Execution Flow

```python
# SQL processes query in this order:

# FROM
# WHERE
# GROUP BY
# HAVING
# SELECT
# ORDER BY
```
