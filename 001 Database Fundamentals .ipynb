```python
# What are databases
# A database is an organized collection of data that is stored electronically and can be easily accessed, managed, and updated.

# Think of it like a smart digital storage system 📦 where data is structured so you can quickly search, filter, and analyze it.
```

```python
# Why are Databases are Important
#1. Store large amounts of data efficiently
#2. Fast retrieval (queries)
#3. Data consistency & accuracy
#4. Multi-user access
#5. Security & backup
#6. Data Analysis
#7. Keeping Records
```

```python
# Properties of Databases
#1. Integrity
    # Ensures data is accurate, correct, and consistent.
    # ex: A student’s age cannot be negative; a foreign key ensures valid references between tables.

#2. Availability
    # Data is accessible whenever authorized users need it.
    # ex: An online banking system is available 24/7 for users to check balances.

#3. Security
    # Protects data from unauthorized access or tampering.
    # Only HR staff can access employee salaries; passwords are encrypted.

#4. Independent of Application
    # Data is stored separately from applications, usable by multiple programs.
    #ex: A company database is used by payroll software, HR software, and analytics tools.

#5. Concurrency
    # Multiple users can access or modify data simultaneously without conflict.
    # ex: Two employees updating different orders in an e-commerce database at the same time.
```

```python
#1. Atomicity
    # A transaction is all or nothing. If any part fails, the entire transaction is rolled back.
    # Transferring money: If $100 is deducted from account A but fails to credit account B, the whole transaction is undone.

#2. Consistency	
    # Transactions take the database from one valid state to another; all rules are enforced.
    # A bank database rule: account balance cannot be negative. A transaction cannot violate this.

#3. Isolation
    # Multiple transactions can occur simultaneously without interfering with each other.
   # Two users updating different products’ stock at the same time won’t affect each other’s changes.

#4. Durability	
    # Once a transaction is committed, it is permanent, even if the system crashes.	
    # After successfully updating a user’s order in e-commerce, the change remains even if the server restarts.
```

```python
# Types of Databases
#1. Relational Databases (RDBMS)
   # Definition: Store data in tables (rows & columns) with relationships between them.
   # Examples: MySQL, PostgreSQL, Oracle, SQL Server
   # Use Case: Banking systems, e-commerce order management

#2. NoSQL Databases
# Definition: 
# Non-relational databases that store data in key-value, document, graph, unstructered or semi-structured data such as images, videos

# Types & Examples:
# Document: MongoDB, CouchDB → store JSON-like documents
# Key-Value: Redis, DynamoDB → fast access using keys
# Graph: Neo4j → stores relationships between entities
# Use Case: Social networks, big data, real-time analytics
```

```python
#3. Columnar Database (Column DB)
# Definition:
# A column-oriented database stores data column by column instead of row by row (like relational databases).

# This is highly optimized for read-heavy operations, especially analytical queries.
# Ideal for large-scale data analytics, data warehousing, and business intelligence.

# Key Features:
# Column Storage: Each column is stored separately on disk.
# Fast Aggregations: Summing, averaging, or filtering a column is much faster.
# Compression: Columns often have similar data, allowing high compression rates.
# Scalable: Handles huge datasets efficiently.

# Examples: Amazon redshift, OLAP, Google BigQuery, Datawarehouse
```

```python
#4. Grph Database
# A graph database stores data as nodes (entities) and edges (relationships) instead of tables.
# Ideal for data with complex relationships.
# Focuses on connections rather than individual records.

# Key Features:
# Nodes: Represent entities (e.g., Person, Product)
# Edges: Represent relationships (e.g., Friend, Bought)
# Properties: Both nodes and edges can have attributes


# Popular Graph Databases
# Neo4j, Amazon Neptune(cloud base db)

# Example
# Nodes: Alice, Bob, Charlie
# Edges: Alice → Friend → Bob, Bob → Friend → Charlie
# Query Example: Find all friends of Alice
```

```python
# Important terms
#1. Attributes/Columns/fields
#2. Rows/records/tuples
#3. Degree of the relation: No of attributes/columns/
#4. Cardinality of the relation: No of rows/records/tuples
#5. Missing value is called null
#6. Domain: A domain in a database is the set of valid values that a column (attribute) can have.
    # It defines the type, format, and allowed range of data.
```

```python
# Keys
# A key is a column (or set of columns) in a table that uniquely identifies a row or defines relationships between tables. 
# Keys ensure data integrity and efficient access.

#1. Primary Key (PK)
# Definition: Uniquely identifies each row in a table.
# Must be unique
# Cannot be NULL


#2. Super Key
# Definition: A set of one or more columns that uniquely identify a row.
# Note: Primary key is a minimal super key.


#3. Candidate Key
# A minimal super key, i.e., no extra column is included; it uniquely identifies a row.
# Note: Each table can have multiple candidate keys, one of which is chosen as the primary key.


#4. Alternate Key
# An alternate key is a candidate key that is not used as the primary key.
# AK = CK - PK
# EX: Primary Key: id , Alternate Key: email  


#5.Composite Key
# Definition: A key made of two or more columns that together uniquely identify a row.
# Together, (student_id, course_id) is the primary key.
```

```python
#6. Foreign Key (FK)
# Definition: A column in one table that refers to the primary key of another table.
# Purpose: Maintains referential integrity between tables.

# CREATE TABLE courses (
#    course_id INT PRIMARY KEY,
#    course_name VARCHAR(50)
#);

#CREATE TABLE enrollments (
#    student_id INT,
#    course_id INT,
#    FOREIGN KEY (course_id) REFERENCES courses(course_id)
#);


#7. Surrogate Key
# A surrogate key is an artificial key (system-generated) used to uniquely identify a row in a table.
# It does not have any business meaning.
# Usually numeric or auto-incremented.
# Often used as the primary key when natural keys are unsuitable.

# CREATE TABLE students (
#    student_id INT AUTO_INCREMENT PRIMARY KEY,  -- surrogate key
#    name VARCHAR(50),
#    email VARCHAR(100)  -- natural key
#);
```

```python
# Cardinality of Relationships
# Cardinality defines the number of entities that can participate in a relationship between tables in a database.
# It specifies how many instances of one entity relate to instances of another entity.


# Types of Cardinality
#1. One-to-One (1:1)
# Each row in Table A relates to exactly one row in Table B, and vice versa.
# Example:  Table Person → Table Passport ,  One person has only one passport.
# Requires only one table for representation


#2. One-to-Many (1:N)
# One row in Table A relates to many rows in Table B.
# Example:
# Table Teacher → Table Students  ,  One teacher can teach many students.
# Requires 2 tables to store the info


#3. Many-to-One (N:1)
# Many rows in Table A relate to one row in Table B.
# Example:
# Table Students → Table Classroom  ,  Many students belong to one classroom.
# # Requires 2 tables to store the info


#4. Many-to-Many (M:N)
# Many rows in Table A relate to many rows in Table B.
# Example:
# Table Students → Table Courses  ,  A student can take many courses, and a course can have many students.
# # Requires 3 tables to store the info
```

```python
# Drawbacks of Databases
#1. Complexity
# Databases are complex to design, implement, and maintain.
# Example: Setting up a relational database with multiple tables, relationships, and constraints requires skilled personnel.


#2. Cost
#High setup, licensing, hardware, and maintenance costs.
# Example: Oracle or SQL Server enterprise editions can be expensive for small businesses.


#3. Scalability
# Some databases may struggle to handle extremely large data or high traffic efficiently.
# Example: A single MySQL server may slow down with millions of transactions per day without sharding or clustering.


#4. Data Integrity
# Corruption or improper transactions can compromise data accuracy.
# Example: If a banking transaction is interrupted, balances may become inconsistent. 


#5. Security
# Databases are vulnerable to unauthorized access, hacking, and data leaks.
# Example: Poorly secured employee database could expose personal information.


#6. Data Migration
# Moving data from one database system to another is time-consuming and error-prone.
# Example: Migrating from MySQL to PostgreSQL may require data type and query adjustments.


#7. Flexibility
# Rigid schema can make it hard to adapt to changing business requirements.
# Example: Adding a new column to a table in a large production database may require downtime or careful planning.
```
