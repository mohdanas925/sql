```python
# Joins

# SQL Joins are used to combine data from multiple tables based on a related column 
# (usually a key like id).
```

```python
# | cid | name  |
# | --- | ----- |
# | 1   | Anas  |
# | 2   | Rahul |


# | order_id | cid |
# | -------- | --- |
# | 101      | 1   |
# | 102      | 2   |
# | 103      | 1   |
```

```python
#1. INNER JOIN
# Returns only matching records from both tables

# SELECT customers.name, orders.order_id
# FROM customers
# INNER JOIN orders
# ON customers.cid = orders.cid;
```

```python
# 2. LEFT JOIN (LEFT OUTER JOIN)

# Returns all records from left table + matched from right
# If no match → NULL

# SELECT customers.name, orders.order_id
# FROM customers
# LEFT JOIN orders
# ON customers.cid = orders.cid;
```

```python
# 3. RIGHT JOIN (RIGHT OUTER JOIN)

# Returns all records from right table + matched from left

# SELECT customers.name, orders.order_id
# FROM customers
# RIGHT JOIN orders
# ON customers.cid = orders.cid;
```

```python
# 4. FULL JOIN (FULL OUTER JOIN)

# Returns all records from both tables
# If no match → NULL

# SELECT customers.name, orders.order_id
# FROM customers
# LEFT JOIN orders ON customers.cid = orders.cid

# UNION

# SELECT customers.name, orders.order_id
# FROM customers
# RIGHT JOIN orders ON customers.cid = orders.cid;
```

```python
#5. SELF JOIN

# A table joins with itself

# SELECT A.name, B.name
# FROM customers A
# JOIN customers B
# ON A.cid = B.cid;
```
