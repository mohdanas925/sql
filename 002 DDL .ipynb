```python
# What is SQL
# SQL (Structured Query Language) is a standard programming language used to store, manage, and retrieve data from databases.
# SQL helps you talk to a database



# What you can do with SQL:

# Create databases and tables
# Insert data
# Retrieve data (queries)
# Update or delete data
# Control access (security)
```

```python
# Types of SQL Commands
# 1. DDL 
# Used to define or change database structure.

# Commands:
# CREATE → create database/table
# ALTER → modify structure
# DROP → delete table/database
# TRUNCATE → remove all data (structure remains)
```

```python
# 2. DML (Data Manipulation Language)
# Used to manage data inside tables.

# Commands:
# INSERT → add data
# UPDATE → modify data
# DELETE → remove data
```

```python
# 3. DQL (Data Query Language)
# Used to retrieve data.

# Command:
# SELECT
```

```python
# 4. DCL (Data Control Language)
# Used to control access/permissions.

# Commands:
# GRANT → give permission
# REVOKE → remove permission
```

```python
# 5. TCL (Transaction Control Language)
# Used to manage transactions (changes in database).

# Commands:
# COMMIT → save changes
# ROLLBACK → undo changes
# SAVEPOINT → temporary save
```

```python
# Data Integrity
# Data integrity in databases refers to the accuracy, completeness and consistency of the data stored in a database.
# It is a measure of realiablity, trustworthiness of the data and ensures that data in a database is protected from errors, corruption and unauthorized changes.


# There are various methods used to ensure data integrity

#1. Constrains
# Constraints are rules applied to table columns to ensure that the data entered into the database is accurate and reliable.
# Purpose: Maintain data integrity by restricting invalid data.


#2. Transactions
# A transaction is a group of SQL operations that are executed as a single unit. Either all succeed or all fail, ensuring data integrity.


#3. Normalization
# Normalization is the process of organizing tables to reduce redundancy and prevent anomalies.
# Purpose: Maintain data integrity by avoiding duplicate or inconsistent data.

# Normal Forms (simplified):

# 1NF: Each column contains atomic values.
# 2NF: Remove partial dependency.
# 3NF: Remove transitive dependency.
# BCNF and higher: Further refinement.
```

```python
# FOREIGN KEY REFERENTIAL ACTIONS
# When a foreign key is linked to a primary key, referential actions define what happens on UPDATE or DELETE.

#1. CASCADE Changes in parent table are automatically reflected in child table

# CREATE TABLE orders(
	# order_id INT PRIMARY KEY AUTO_INCREMENT,
    # cid INT NOT NULL,
    # order_date DATETIME NOT NULL DEFAULT CURRENT_TIMESTAMP,
    
    # CONSTRAINT order_fk FOREIGN KEY (cid) REFERENCES customers (cid)

    # ON DELETE CASCADE
    # ON UPDATE CASCADE
# );
```

```python
#2. SET NULL Child foreign key becomes NULL when parent is deleted/updated.

# CREATE TABLE orders(
	# order_id INT PRIMARY KEY AUTO_INCREMENT,
    # cid INT NOT NULL,
    # order_date DATETIME NOT NULL DEFAULT CURRENT_TIMESTAMP,
    
    # CONSTRAINT order_fk FOREIGN KEY (cid) REFERENCES customers (cid)
    
    # ON DELETE SET NULL
    # ON UPDATE SET NULL
    # -- SET NULL requires the column to allow NULL values.
# );
```

```python
#3. RESTRICT -> Prevents delete/update if child rows exists

# CREATE TABLE orders(
	# order_id INT PRIMARY KEY AUTO_INCREMENT,
    # cid INT NOT NULL,
    # order_date DATETIME NOT NULL DEFAULT CURRENT_TIMESTAMP,
    
    # CONSTRAINT order_fk FOREIGN KEY (cid) REFERENCES customers (cid)
    
    # ON DELETE RESTRICT
    # ON UPDATE RESTRICT
# );
```
