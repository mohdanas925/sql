```python
# Types of Sub-queries
# 1. Based on the result they returns

# a). Scalar Subquery
# A scalar subquery returns exactly one value (1 row, 1 column)
# It is used where a single value is expected (like =, >, <, etc.)


# SELECT name
# FROM students
# WHERE marks = (SELECT MAX(marks) FROM students);
```
